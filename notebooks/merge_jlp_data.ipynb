```python
#rename all column headers except date by adding "GT" to name
terminal_jlp_sol.columns = terminal_jlp_sol.columns.map(lambda x: 'GT_' + x if x != 'date' else x)
```

```python
#add "usd" to end of column header
terminal_jlp_sol.columns = terminal_jlp_sol.columns.map(lambda x: x + '_usd' if x == 'GT_volume' else x)
```

```python
#drop columns GT_24h_volume, GT_6h_volume, GT_1h_volume, GT_5m_volume
terminal_jlp_sol.drop(['GT_24h_volume', 'GT_6h_volume', 'GT_1h_volume', 'GT_5m_volume'], axis=1, inplace=True)
```

#CLEANING JLP_SOL_CG

```python
coin_jlp_sol.head()
```

```python
#rearrage by putting most recent date first
coin_jlp_sol = coin_jlp_sol.iloc[::-1]
```

#MERGE DATA

```python
terminal_jlp_reset = terminal_jlp_sol.reset_index()

coin_jlp_reset = coin_jlp_sol.reset_index()
```

```python
merged_jlp_sol = pd.merge(
    terminal_jlp_reset,
    coin_jlp_reset,
    on='date',
    how='outer'  # keeps all dates from both DataFrames
)
```

```python
#remove index_x and index_y
merged_jlp_sol.drop(['index_x', 'index_y'], axis=1, inplace=True)
```

```python
merged_jlp_sol = merged_jlp_sol.iloc[::-1]
```

```python
merged_jlp_sol.info()
```

```python
#save and download as csv
merged_jlp_sol.to_csv('merged_jlp_sol.csv', index=False)
```
